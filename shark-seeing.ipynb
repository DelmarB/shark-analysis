```python
# Ensure all columns are visible
pd.set_option('display.max_columns', None)

# Filter for Australia (case-insensitive)
australia_shark_df = shark_df[shark_df['Country'].str.upper() == 'AUSTRALIA']

# Print row count
row_count = australia_shark_df.shape[0]
print(f"Total shark incidents in Australia: {row_count}")

# Keep only specific columns
columns_to_keep = [
    'Date', 'Year', 'Location', 'Activity', 'Injury',
    'Fatal (Y/N)', 'Species', 'Case Number', 'Case Number.1'
]
available_columns = [col for col in columns_to_keep if col in australia_shark_df.columns]
australia_shark_df = australia_shark_df[available_columns]

# Sort by Location
sorted_aus_shark_df = australia_shark_df.sort_values(by='Location')

# Display table (this is your formatted table)
display(sorted_aus_shark_df)

# Get unique locations and print them
locations_df = sorted_aus_shark_df['Location'].dropna().unique()

print("\nUnique shark incident locations in Australia:")
for location in locations_df:
    print(location)
```
